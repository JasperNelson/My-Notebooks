```python
import numpy as np
import matplotlib.pyplot as plt
import openpyxl as open
import pandas as pd
# Generate a random normal distribution
mu, sigma = 100, 5 # mean and standard deviation
data = np.random.normal(mu, sigma, 100000)
df = pd.DataFrame(data, columns=['Data'])
df.to_excel('give some path', index=False)
```

```python
# Plot the histogram
plt.hist(data, bins=300, alpha=0.5, color='blue', edgecolor='black')
plt.title('Histogram of a Randomly Generated Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
def zscored(datalist):
    for x in range(len(datalist)):
        datalist[x]=((datalist[x])-mu)/sigma
    return datalist
data =zscored(data)
plt.hist(data, bins=300, alpha=0.5, color='pink', edgecolor='red')
plt.title('Histogram of every element in previous graph zscored')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()

```
